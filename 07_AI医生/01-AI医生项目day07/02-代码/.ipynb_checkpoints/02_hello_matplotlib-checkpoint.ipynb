# hello matplotlib

```python
# 1. 导包
import matplotlib.pyplot as plt
# 2. 绘制画布
plt.figure(figsize=(20, 8), dpi=100)
# 3. 准备数据
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
# 4. 绘图
plt.plot(x, y)
# 5. 显示
plt.show()
```

# 设置画布属性与图片保存

```python
# 2. 绘制画布
plt.figure(figsize=(20, 8), dpi=100)
# 3. 准备数据
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
# 4. 绘图
plt.plot(x, y)
# 保存图像 
# 注意：保存图像要放在show()函数之前，因为show之后会释放计算机的内存资源
plt.savefig("test.png")
# 5. 显示
plt.show()
```

# 辅助显示层

```python
# 1. 导包
import random
# 2. 准备数据
x = range(60)
y_shenzen = [random.uniform(26,38) for i in x]

# 3. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 4. 绘图
plt.plot(x, y_shenzen)

# 构造x轴标签
x_ticks_label = ['11点{}分'.format(i) for i in x ]
# 构造y轴标签
y_ticks = range(40)

plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks[::5])

# 5. 显示
plt.show()
```
